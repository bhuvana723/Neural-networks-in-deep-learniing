```python
from sklearn.model_selection import GridSearchCV
from scikeras.wrappers import KerasClassifier  # Changed import statement

# Define a function to create your model
def create_model(neurons=64, activation='relu', learning_rate=0.01):
    model = keras.Sequential([
        keras.layers.Dense(neurons, activation=activation, input_shape=(number_of_input_features,)),
        keras.layers.Dense(number_of_output_classes, activation='softmax')
    ])
    optimizer = keras.optimizers.Adam(learning_rate=learning_rate)
    model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Create a KerasClassifier wrapper for your model
model = KerasClassifier(model=create_model, verbose=0) # Updated parameter name to 'model'

# Define the hyperparameter grid
# Pass hyperparameters as arguments to create_model function
param_grid = {
    'model__neurons': [32, 64, 128],  # Notice 'model__' prefix
    'model__activation': ['relu', 'tanh'],  # Notice 'model__' prefix
    'model__learning_rate': [0.01, 0.001, 0.0001]  # Notice 'model__' prefix
}

# Create GridSearchCV object
grid = GridSearchCV(estimator=model, param_grid=param_grid, n_jobs=-1, cv=3)

# Fit the grid search to your data
grid_result = grid.fit(X_train, y_train)

# Print the best parameters and score
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```
